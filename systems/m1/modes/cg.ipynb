```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
```

```python
data = np.load("cg_data_m1s1m6.pkl",allow_pickle=True)
xy = np.asarray(data['xy']).reshape(-1,2)
z = data["modal_surface"]
z_cg = data["cg_surface"]
```

```python
fig,ax = plt.subplots()
ax.triplot(xy[:,0],xy[:,1])
ax.set_aspect("equal")
ax.grid
```

```python
fig,ax = plt.subplots()
h = ax.tripcolor(xy[:,0],xy[:,1],z)
ax.set_aspect("equal")
ax.grid()
fig.colorbar(h,ax=ax)
```

```python
fig,ax = plt.subplots()
h = ax.tripcolor(xy[:,0],xy[:,1],z_cg)
ax.set_aspect("equal")
ax.grid()
fig.colorbar(h,ax=ax)
```
