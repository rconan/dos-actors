```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = 40
d = 2/100
a = 2*np.pi*f
b = 2*a*d
T = 1e-3
tau = 1e-3
g = 0.75
```

```python
fig,ax = plt.subplots()
sys = np.load("mount-el-tf.pkl",allow_pickle=True)
nu = sys[0]
tf = 20*np.log10(sys[1])
ax.semilogx(nu,tf,label="mount elevation axis")
sys = np.load("mount-az-tf.pkl",allow_pickle=True)
nu = sys[0]
tf = 20*np.log10(sys[1])
ax.semilogx(nu,tf,label="mount azimuth axis")
sys = np.load("mount-rot-tf.pkl",allow_pickle=True)
nu = sys[0]
tf = 20*np.log10(sys[1])
ax.semilogx(nu,tf,label="mount rotator axis")
ax.grid()
s = 2*np.pi*1j*np.asarray(nu)
G = g*np.exp(-tau*s)*(1-np.exp(-T*s))/(T*s)**2
ax.semilogx(nu,20*np.log10(np.abs(1/(1+G)))-200,'k',alpha=0.25,label="AO RTF (1kHz,2ms,0.75)")

ax.set_xlabel("Frequency [Hz]")
ax.set_ylabel("Transfer Function [dB]")
ax.legend()
```
