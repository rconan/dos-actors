```python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('dark_background')
%matplotlib inline
```

```python
(nu,fr) = np.load("asm_kl_fr.pkl",allow_pickle=True)
```

```python
fr = np.asarray(fr)
tf = fr[...,0] + 1j*fr[...,1]
tf_db = 20*np.log10(np.abs(tf))
```

```python
def bodemag(ax) :
    ax.grid(which='both')
    ax.set_xlabel('Frequency [Hz]')
    ax.set_ylabel("ASM#1 KL mode PSDs [dB]")
    ax.set_xlim([10,4e3])
    ax.set_ylim([-20,10])
    ax.set_yticks([-20,-15,-10,-3,0,-6,6,10])
    ax.fill_between([10,4e3],[-3, -3],6,color="C1",alpha=0.25)
    ax.fill_between([800,4e3],[-20, -20],-3,color="C1",alpha=0.25)
```

```python
data = np.load("../../white-noise/ASMS_TF.npz")
wn_nu = data['nu']
wn_tf_db = data['tf_db']
```

```python
fig,axs = plt.subplots(nrows=4,ncols=2,sharex=True,sharey=True,figsize=(20,20))
ax =axs[0,0]
ax.semilogx(nu,tf_db[:,0],label="piston")
ax.semilogx(nu,tf_db[:,5],label="focus")
ax.legend()
bodemag(ax)
ax =axs[1,0]
ax.semilogx(nu,tf_db[:,:6])
ax.legend(list(range(1,tf_db.shape[0]+1)))
bodemag(ax)
ax =axs[2,0]
ax.semilogx(nu,tf_db[:,:66],"C0",alpha=0.25)
ax.set_title(" 1st 66 modes")
bodemag(ax)
ax =axs[3,0]
ax.semilogx(nu,tf_db,"C0,")
ax.set_title("496 modes")
bodemag(ax)

ax =axs[0,1]
ax.semilogx(wn_nu,wn_tf_db[:,0],label="piston")
ax.semilogx(wn_nu,wn_tf_db[:,5],label="focus")
ax.legend()
bodemag(ax)
ax =axs[1,1]
ax.semilogx(wn_nu,wn_tf_db[:,:6])
ax.legend(list(range(1,wn_tf_db.shape[0]+1)))
bodemag(ax)
ax =axs[2,1]
ax.semilogx(wn_nu,wn_tf_db[:,:66],"C0",alpha=0.25)
ax.set_title(" 1st 66 modes")
bodemag(ax)
ax =axs[3,1]
ax.semilogx(wn_nu,wn_tf_db,"C0,")
ax.set_title("496 modes")
bodemag(ax)
```
