```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(nu,tf_rs) = np.load("mount_el_tf_dc.pkl",allow_pickle=True)
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
tfm_dc = np.mean(tf,(0,1))
```

```python
(nu,tf_rs) = np.load("mount_el_tf_dc_half-sampling-delay.pkl",allow_pickle=True)
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
tfm_dch = np.mean(tf,(0,1))
```

```python
(nu,tf_rs) = np.load("mount_el_tf_dc_full-sampling-delay.pkl",allow_pickle=True)
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
tfm_dcf = np.mean(tf,(0,1))
```

```python
(nu,tf_rs) = np.load("mount_el_tf.pkl",allow_pickle=True)
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
tfm = np.mean(tf,(0,1))
fig,ax = plt.subplots()
ax.semilogx(nu,20*np.log10(np.abs(tfm)))
ax.semilogx(nu,20*np.log10(np.abs(tfm_dc)))
ax.semilogx(nu,20*np.log10(np.abs(tfm_dch)))
ax.semilogx(nu,20*np.log10(np.abs(tfm_dcf)))
ax.grid(which="both")
ax.set_xlabel("frequency [Hz]")
ax.set_label("Avg El Enc")
```

```python
(nu,tf_rs) = np.load("mount_el_tf.pkl",allow_pickle=True)
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
tfm = np.mean(tf,(0,1))
fig,ax = plt.subplots()
ax.semilogx(nu,np.angle(tfm),label="w/o DC comp.")
ax.semilogx(nu,np.angle(tfm_dc),label="w/ DC comp. (no delay)")
ax.semilogx(nu,np.angle(tfm_dch),label="w/ DC comp. (half sampling delay)")
ax.semilogx(nu,np.angle(tfm_dcf),label="w/ DC comp. (full sampling delay)")
ax.grid(which="both")
ax.set_xlabel("frequency [Hz]")
ax.set_label("Avg El Enc")
ax.legend()
```
