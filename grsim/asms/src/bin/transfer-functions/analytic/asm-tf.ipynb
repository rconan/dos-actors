```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(nu,tf_rs) = np.load("asm_tf_100log.pkl",allow_pickle=True)
```

```python
n = tf_rs[0][1]
m = tf_rs[0][2]
n,m
```

```python
res = [list(zip(*x[0])) for x in tf_rs]
tf = np.dstack([np.asarray(x[0]).reshape(n,m,order="F") for x in res]) \
     +1j*np.dstack([np.asarray(x[1]).reshape(n,m,order="F") for x in res])
```

```python
tf.shape
```

```python
fig,ax = plt.subplots()
ax.semilogx(nu,20*np.log10(np.abs(tf[0,0,:])))
ax.semilogx(nu,20*np.log10(np.abs(tf[1,1,:])))
ax.semilogx(nu,20*np.log10(np.abs(tf[2,2,:])))
ax.semilogx(nu,20*np.log10(np.abs(tf[3,3,:])))

ax.semilogx(nu,20*np.log10(np.abs(tf[5,5,:])))
ax.grid(which="both")
ax.set_xlabel("frequency [Hz]")
ax.set_label("Avg El Enc")
```

```python
fig,ax = plt.subplots()
ax.semilogx(nu,20*np.log10(np.abs(tf[416,416,:])))
ax.grid(which="both")
ax.set_xlabel("frequency [Hz]")
ax.set_label("Avg El Enc")
```
