```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.tri as tri
%matplotlib inline
```

```python
def load(var):
    data = np.load(var, allow_pickle=True)
    return np.asarray(data[0],order="F").reshape(data[1],data[2])
```

```python
vc_f2d = load("vc_f2d.pkl")
fs_f2d = load("fs_f2d.pkl")
```
