```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_parquet("model-data_1.parquet")
t = np.arange(len(df))/8000
df
```

```python
asms1_kl_coeff = np.vstack(df["FaceSheetFigure#1"])
asms2_kl_coeff = np.vstack(df["FaceSheetFigure#2"])
asms5_kl_coeff = np.vstack(df["FaceSheetFigure#5"])
asms6_kl_coeff = np.vstack(df["FaceSheetFigure#6"])
asms7_kl_coeff = np.vstack(df["FaceSheetFigure#7"])
```

```python
fig,axs = plt.subplots(nrows=2, ncols=3, figsize=(12,7), sharex=True)
ax = axs[0,0]
ax.plot(t,asms1_kl_coeff[:,:100])
# ax.hlines(one_arcsec,0,8,'k')
ax.set_ylabel("ASM-S1 KL modal coeff"), ax.grid()
ax.set_xlabel("Time [s]")

ax = axs[0,1]
ax.plot(t,asms2_kl_coeff[:,:100])
# ax.hlines(one_arcsec,0,8,'k')
ax.set_ylabel("ASM-S2 KL modal coeff"), ax.grid()
ax.set_xlabel("Time [s]")

ax = axs[1,0]
ax.plot(t,asms5_kl_coeff[:,:100])
# ax.hlines(one_arcsec,0,8,'k')
ax.set_ylabel("ASM-S5 KL modal coeff"), ax.grid()
ax.set_xlabel("Time [s]")

ax = axs[1,1]
ax.plot(t,asms6_kl_coeff[:,:100])
ax.set_ylabel("ASM-S6 KL modal coeff"), ax.grid()
ax.set_xlabel("Time [s]")

ax = axs[1,2]
ax.plot(t,asms7_kl_coeff[:,:100])
ax.set_ylabel("ASM-S7 KL modal coeff"), ax.grid()
ax.set_xlabel("Time [s]")
```
