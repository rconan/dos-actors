```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_parquet("elevation.parquet")
t = np.arange(len(df))/1000
encoders = np.vstack(df["MountEncoders"])
one_arcsec = np.pi/180/3600

fig,axs = plt.subplots(nrows=3,figsize=(7,12),sharex=True)
ax = axs[0]
ax.plot(t,encoders[:,4:10].mean(1));
ax.hlines(one_arcsec,0,8,'k')
ax.grid()
ax.set_ylabel("Encoders mean [rd]")
ax.set_title("Elevation")
ax = axs[1]
ax.plot(t,encoders[:,:4].mean(1));
#ax.hlines(one_arcsec,0,8,'k')
ax.grid()
ax.set_ylabel("Encoders mean [rd]")
ax.set_title("Azimuth")
ax = axs[2]
ax.plot(t,encoders[:,10:].mean(1));
#ax.hlines(one_arcsec,0,8,'k')
ax.grid()
ax.set_title("GIR")
ax.set_xlabel("Time [s]")
ax.set_ylabel("Encoders mean [rd]")
```
